# Some more important points : loc and iloc based selection.

```python
#syntax: .loc is used for dataset selection based on column names, whereas .iloc is used for dataset selection based on indexes.
```

```python
# syntax: .loc[<row_selection>,<col_selection>]

# <row_selection> = first row |  last row | one  row | all rows | some rows based on condition | list of rows | range of rows.

# <col_selection> = first col | last col | one col | all cols | some columns based on condition | list of cols | range of cols.
```

```python
# first row == df.iloc[0]
# last row == df.iloc[-1]
# one row == df.iloc[4]
# all rows == df[:]
# some rows based on condition == df.iloc[ds['age']>23]
# all rows having name starts with a and contain r. == df.iloc[df['name'].str.startswith('a') & df['name'].str.contains('r')]
# list of rows selection == df.iloc[[0,1,2,2,3]]
# range of rows selection == df.iloc[2:5]


# same can be done for columns as well.

# first col = df.iloc[,0]
# last col =df.iloc[,-1]
# one col =df.iloc[,3]
# all cols=df.iloc[,:]
# some cols based on condition = 
# cols_to_be_selected=[True, False , True , False , True , True, False]
# cols_final= [x for x in range(len(cols_to_be_selected)) if cols_to_be_selected[x]==True]


# list of cols = df.iloc[,[1,2,3]]
# range of cols=df.iloc[,2:4]


```
